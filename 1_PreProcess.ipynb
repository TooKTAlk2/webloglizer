## 날짜 분리 및 시각(sec)만들기
- 날짜를 day,hour,minute,second,AMPM으로 분리
- 분리된 날짜를 첫 인덱스(8:24 10:00:18 AM == 0)으로 시각을 Time 칼럼 sec단위로 만듦

```python
data["Day"] = 0
data["Hour"] = 0
data["Minute"] = 0
data["Second"] = 0
data["AMPM"] = 0

for i in data.index:
    time = data.loc[i,"Timestamp"]
    split_time = time.split(' ')
    split_time[0].split('.')
    y,m,d = map(int,split_time[0].split('.'))
    h,minute,sec = map(int,split_time[1].split(':'))
    ampm = split_time[2]
    data.loc[i,"Day"] = d
    data.loc[i,"Hour"] = h
    data.loc[i,"Minute"] = minute
    data.loc[i,"Second"] = sec
    data.loc[i,"AMPM"] = ampm
    
```

```python
# 기준은 첫 index  8:24 10:00:18
start_d = data.iloc[0,10]
start_h = data.iloc[0,11]
start_m = data.iloc[0,12]
start_s = data.iloc[0,13]
data["Time"] = 0
for i in data.index:
    # data의 AM/PM와 hour를 이용하여 12시간 표기법을 24시 표기법으로 변환
    if data.loc[i,"AMPM"] == "PM":
        data.loc[i,"Hour"] = data.loc[i,"Hour"] + 12
    
    d = data.iloc[i,10]
    h = data.iloc[i,11]
    m = data.iloc[i,12]
    s = data.iloc[i,13]
    # 무조건 양수,
    rest_d = d - start_d
    rest_h = h - start_h
    rest_m = m - start_m
    rest_s = s - start_s
    rest_d_to_s = rest_d * 24 * 60 * 60
    rest_h_to_s = rest_h * 60 * 60
    rest_m_to_s = rest_m * 60
    data.loc[i,"Time"] = rest_d_to_s + rest_h_to_s + rest_m_to_s + rest_s
```

```python
data.to_csv("./created_time_sample.csv",index=False)
```
